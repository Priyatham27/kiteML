# KiteML Tutorial: Pipeline & Model Deployment

This notebook demonstrates building, saving (.kml package), and serving models with **KiteML**.

```python
import pandas as pd
from kiteml import KiteMLPipeline

df = pd.DataFrame({
    "age": [25, 45, 35, 50, 23, 62, 48, 29],
    "income": [50000, 90000, 62000, 110000, 48000, 125000, 98000, 54000],
    "target": [0, 1, 0, 1, 0, 1, 1, 0]
})

pipeline = KiteMLPipeline()
build_res = pipeline.build(df, target="target")

# Save as .kml binary package
pipeline.save("my_deployment_pipeline.kml")
print("Successfully saved .kml deployment package!")
```
