# KiteML Tutorial: Multiclass Classification

This notebook demonstrates multiclass classification training with **KiteML**.

```python
import pandas as pd
from kiteml import train

# Create synthetic multiclass dataset
df = pd.DataFrame({
    "feature1": [1.2, 2.4, 3.1, 4.5, 1.1, 2.9, 3.8, 4.1],
    "feature2": [5.1, 4.2, 3.3, 2.1, 5.0, 4.1, 3.0, 2.2],
    "category": ["class_A", "class_B", "class_C", "class_C", "class_A", "class_B", "class_C", "class_B"]
})

result = train(data=df, target="category", problem_type="classification")
print(result.summary())
```
